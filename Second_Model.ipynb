# Handling Outliers in the Dataset

> -  ##### `first`, I explore Ranges for each Feature 
> -  ##### `second`, aggregate every feature with those similar in range 
> -  ##### `third`, I make a threshold for every group of Features 
> -  ##### `finally`, trans big Ranges to small numbers to handle outliers


```python
print(f"{"="*20} Column Ranges {"="*20}")

def column_ranges():
    for column in df.columns:
        column_range = int(df[column].max() - df[column].min())
        print(f"{column} = {column_range:,}")


column_ranges()
```

```python
columns = ["BALANCE","PURCHASES","ONEOFF_PURCHASES","INSTALLMENTS_PURCHASES","CASH_ADVANCE","CREDIT_LIMIT","PAYMENTS","MINIMUM_PAYMENTS"]


for col in columns:
    Range=col +'_RANGE'
    df[Range]=0        
    df.loc[((df[col]>0)&(df[col]<=500)),Range]=1
    df.loc[((df[col]>500)&(df[col]<=1000)),Range]=2
    df.loc[((df[col]>1000)&(df[col]<=3000)),Range]=3
    df.loc[((df[col]>3000)&(df[col]<=5000)),Range]=4
    df.loc[((df[col]>5000)&(df[col]<=10000)),Range]=5
    df.loc[((df[col]>10000)),Range]=6

```

```python
columns = ["BALANCE_FREQUENCY","PURCHASES_FREQUENCY","ONEOFF_PURCHASES_FREQUENCY","PURCHASES_INSTALLMENTS_FREQUENCY","CASH_ADVANCE_FREQUENCY","PRC_FULL_PAYMENT"]


for col in columns:

    
    
    Range=col+'_RANGE'
    df[Range]=0
    df.loc[((df[col]>0)&(df[col]<=0.1)),Range]=1
    df.loc[((df[col]>0.1)&(df[col]<=0.2)),Range]=2
    df.loc[((df[col]>0.2)&(df[col]<=0.3)),Range]=3
    df.loc[((df[col]>0.3)&(df[col]<=0.4)),Range]=4
    df.loc[((df[col]>0.4)&(df[col]<=0.5)),Range]=5
    df.loc[((df[col]>0.5)&(df[col]<=0.6)),Range]=6
    df.loc[((df[col]>0.6)&(df[col]<=0.7)),Range]=7
    df.loc[((df[col]>0.7)&(df[col]<=0.8)),Range]=8
    df.loc[((df[col]>0.8)&(df[col]<=0.9)),Range]=9
    df.loc[((df[col]>0.9)&(df[col]<=1.0)),Range]=10
```

```python
column = ["PURCHASES_TRX","CASH_ADVANCE_TRX"]

for col in column:

    Range=col +'_RANGE'
    df[Range]=0
    df.loc[((df[col]>0)&(df[col]<=5)),Range]=1
    df.loc[((df[col]>5)&(df[col]<=10)),Range]=2
    df.loc[((df[col]>10)&(df[col]<=15)),Range]=3
    df.loc[((df[col]>15)&(df[col]<=20)),Range]=4
    df.loc[((df[col]>20)&(df[col]<=30)),Range]=5
    df.loc[((df[col]>30)&(df[col]<=50)),Range]=6
    df.loc[((df[col]>50)&(df[col]<=100)),Range]=7
    df.loc[((df[col]>100)),Range]=8


```

# Drop old Features with it's outliers

```python
colmns = ["PURCHASES_TRX","CASH_ADVANCE_TRX","BALANCE_FREQUENCY","PURCHASES_FREQUENCY","ONEOFF_PURCHASES_FREQUENCY","PURCHASES_INSTALLMENTS_FREQUENCY","CASH_ADVANCE_FREQUENCY","PRC_FULL_PAYMENT","BALANCE","PURCHASES","ONEOFF_PURCHASES","INSTALLMENTS_PURCHASES","CASH_ADVANCE","CREDIT_LIMIT","PAYMENTS","MINIMUM_PAYMENTS"]
df.drop(columns=colmns,inplace=True)
```

# Scaling data before pass it to the model

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
new_df = std_scaler.fit_transform(df)
```

```python
pd.DataFrame(new_df)
```

# Elbow Method to choose Best `k`

```python
from sklearn.cluster import KMeans 

sentroids = []
for k in range(3,31):
    algo = KMeans(n_clusters=k)
    algo.fit(new_df)
    sentroids.append(algo.inertia_)
```

```python

plt.figure(figsize=(6,6))
sns.lineplot(sentroids)
plt.title("elbow Method to choose suitable (K) for K-Means")
plt.xlabel("(K) number of Clusters")
plt.ylabel("Within-Cluster Sum of Squares (WCSS)")
plt.grid()
```

# Passing Data to Kmeans algorithm 

```python
n_clusters=7
model = KMeans(n_clusters=n_clusters,random_state=42,n_init=20)
model.fit(new_df)
labels = model.labels_
sentroids = model.cluster_centers_
```

```python
clusters=pd.concat([df, pd.DataFrame({'cluster':labels})], axis=1)
clusters.head()
```

# Reduce Dimensionality Data  to visualise it

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import PCA 

dist = cosine_similarity(new_df)
pca_model = PCA(n_components=2)
PCA_data = pca_model.fit_transform(dist)


```

```python
pd.DataFrame(PCA_data)
```

# Final Result

```python
plt.figure(figsize=(20,13))
plt.scatter(PCA_data[:,0], PCA_data[:,1], marker='o',c=labels)
plt.scatter(sentroids[:,0], sentroids[:,1], marker='^',color = "red",s=200,label="Centroids")
plt.legend(frameon=True,ncol=1,fancybox=True,framealpha=0.5,shadow=True,borderpad=1,fontsize=15)
plt.colorbar()
plt.title("Customers Segmentation based on their Credit Card usage bhaviour.")
plt.show()
```
